```python
# Import Libraries
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Dataset
X = 1.5
y = 0.5
```

```python
# Initialize a random weight
W = np.random.randn(1)
print(W, W.shape)
```

```python
# Build the training loop
epochs = 10
lr = 0.1
eList = []
for epoch in range(epochs):
    # Forward Pass
    y_out = X*W

    # Compute the error
    error = np.square(y-y_out)
    eList.append(error)

    # Compute the gradient
    gradient = -2*(y-y_out)*X

    # Backpropagate the error
    W = W - gradient*lr
```

```python
# Plot the Epochs vs Error plot
plt.plot(range(epochs), eList)
plt.xlabel("Epochs")
plt.ylabel("Error")
plt.title("Epochs vs Error")
```

```python
# Function for Computing output
def output(X,W):
    return X*W

print(output(X,W))
```
