```python
# Import Libraries
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Dataset
X = np.array([[1.5, 0.625]])
y = 0.5
```

```python
# Initialize random weights
W = np.random.rand(1,2)
```

```python
# Build the training loop
epochs = 20
lr = 0.1
eList = []
for epoch in range(epochs):
    # Forward Pass
    y_out = np.dot(X,np.transpose(W))

    # Compute the error
    error = np.square(y-y_out)
    eList.append(error[0][0])

    # Compute the gradient
    gradient = -2*(y-y_out)*X

    # Backpropagate the error
    W = W - gradient*lr
```

```python
# Build the training loop
plt.plot(range(epochs), eList)
plt.xlabel("Epochs")
plt.ylabel("Error")
plt.title("Epochs vs Error")
```

```python
# Function for Computing output
def output(X, W):
    return np.dot(X, np.transpose(W))

print(output(X,W))
```

```python
# Display the final weight
print(W)
```
